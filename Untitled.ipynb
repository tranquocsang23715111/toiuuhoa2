```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# =====================
# DỮ LIỆU
# =====================
A = np.array([[1, 0],
              [0, -2]], dtype=float)
b = np.array([2, 4], dtype=float)

# =====================
# HÀM f, GRADIENT
# =====================
def f(x):
    x1, x2 = x
    term1 = np.exp(2 * x1)
    term2 = 0.5 * np.sum((A @ x - b)**2)
    return term1 + term2

def grad_f(x):
    x1, x2 = x
    g1 = 2 * np.exp(2 * x1) + (x1 - 2)
    g2 = 8 * x2 + 16
    return np.array([g1, g2])

# =====================
# a) TÌM NGHIỆM TỐI ƯU x*
# =====================
def eq_x1(x1):
    return 2 * np.exp(2 * x1) + (x1 - 2)

x1_star = fsolve(eq_x1, 0)[0]
x2_star = -2
x_star = np.array([x1_star, x2_star])
p_star = f(x_star)

print("Điểm tối ưu x* =", x_star)
print("Giá trị tối ưu p* =", p_star)

# =====================
# b) GRADIENT DESCENT
# =====================

def objective_function(x):
    return np.exp(2*x[0]) + 0.5 * np.linalg.norm(A @ x - b)**2

# ===== Gradient =====
def gradient(x):
    g1 = 2*np.exp(2*x[0]) + (A.T @ (A @ x - b))[0]
    g2 = (A.T @ (A @ x - b))[1]
    return np.array([g1, g2])

# ===== Thuật toán Gradient Descent =====
def gradient_descent(learning_rate, max_iterations, init_x):
    x = init_x.copy()
    x_values = [x]
    f_values = [objective_function(x)]

    for i in range(max_iterations):
        x = x - learning_rate * gradient(x)
        x_values.append(x)
        f_values.append(objective_function(x))
        print(f'Lần lặp {i+1:02d}: x = {x}, f(x) = {f_values[-1]:.6f}')
    return np.array(x_values), np.array(f_values)

# ===== Thiết lập tham số =====
x0 = np.array([-3.0, 5.0])
max_iter = 100

# Thử 2 learning rate khác nhau
learning_rates = [2/5, 1/2]

results = {}

for lr in learning_rates:
    print(f"\n=== Learning rate = {lr} ===")
    x_vals, f_vals = gradient_descent(lr, max_iter, x0)
    results[lr] = (x_vals, f_vals)

# ===== Vẽ biểu đồ hội tụ f(x) =====
plt.figure(figsize=(7,4))
for lr, (x_vals, f_vals) in results.items():
    plt.plot(f_vals, label=f'lr = {lr}')
plt.xlabel('Số vòng lặp (k)')
plt.ylabel('Giá trị f(x)')
plt.title('Quá trình hội tụ của Gradient Descent')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar

# Dữ liệu
A = np.array([
    [0, -2, -1, -2],
    [-1, -1, -3, 0],
    [1, -1, 2, -5]
])
b = np.array([3, -1, 3, 2])
c = np.array([4, 11/4, 13/4])

# Hàm mục tiêu
def f(x):
    val = b - A.T @ x
    # Nếu có phần tử <= 0 → ngoài miền xác định
    if np.any(val <= 0):
        return np.inf
    return c @ x - np.sum(np.log(val))

# Gradient
def grad(x):
    val = b - A.T @ x
    if np.any(val <= 0):
        return np.ones_like(x) * np.inf
    return c + np.sum(A / val, axis=1)

# Gradient Descent với Exact Line Search
def exact_line(x, gradient):
    # Chỉ tìm trong khoảng nhỏ để không vượt miền
    grad_f = lambda learning_rate: f(x - learning_rate * gradient)
    result = minimize_scalar(grad_f, bounds=(0, 0.05), method='bounded')
    return result.x
    
def gradient_descent_exact_line(max_iter, init_x):
    x = init_x
    xs = [x]
    fs = [f(x)]

    for i in range(max_iter):
        g = grad(x)
        if np.any(np.isinf(g)):
            break

        learning_rate = exact_line(x, g)
        new_x = x - learning_rate * g

        # Nếu bước đi vượt ra ngoài miền log → dừng
        if np.any(b - A.T @ new_x <= 0):
            break

        x = new_x
        xs.append(x)
        fs.append(f(x))

        # Dừng nếu gradient đủ nhỏ
        if np.linalg.norm(g) < 1e-6:
            break

    return xs, fs

# Chạy thử
x0 = np.array([-1, 1, 0])
xs, fs = gradient_descent_exact_line(max_iter=100, init_x=x0)

# Kết quả
print("Số bước lặp:", len(fs))
print("Giá trị f(x) nhỏ nhất đạt được:", fs[-1])
print("x* xấp xỉ =", xs[-1])

# Vẽ đồ thị hội tụ
plt.plot(fs, marker='o')
plt.xlabel('Iteration')
plt.ylabel('f(x)')
plt.title('Quá trình hội tụ của Gradient Descent (Exact Line Search)')
plt.grid(True)
plt.show()
```
